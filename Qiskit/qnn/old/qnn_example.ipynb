```python
from qiskit import Aer, QuantumCircuit, transpile, assemble
from qiskit.visualization import plot_histogram
import numpy as np

# Quantum neural network circuit
def quantum_neural_network(input_data, weights):
    # Define the quantum circuit
    num_qubits = len(input_data)
    quantum_circuit = QuantumCircuit(num_qubits)

    # Apply Hadamard gates on input qubits
    quantum_circuit.h(range(num_qubits))

    # Apply parametrized rotations based on weights
    for i in range(num_qubits):
        quantum_circuit.ry(weights[i], i)

    # Measurement
    quantum_circuit.measure_all()

    print(quantum_circuit)
    # Simulate the quantum circuit
    simulator = Aer.get_backend('qasm_simulator')
    transpiled_circuit = transpile(quantum_circuit, simulator)
    result = simulator.run(transpiled_circuit).result()
    counts = result.get_counts()

    return counts

# Generate random input data and weights
input_data = np.random.rand(3)
weights = np.random.rand(3)

# Run the quantum neural network
output_counts = quantum_neural_network(input_data, weights)

# Display the results
print("Input Data:", input_data)
print("Weights:", weights)
print("Output Counts:", output_counts)

# Plot the histogram of the output counts
plot_histogram(output_counts)
```
